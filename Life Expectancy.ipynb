```python
# outlier detection

for i in life.columns:
    q1=np.quantile(life[i],.25)
    q3=np.quantile(life[i],0.75)
    iqr=q3-q1
    ub=q3+(1.5*iqr)
    lb=q1-(1.5*iqr)
```

```python
outliers=life[(life[i]>ub)|(life[i]<lb)]
```

```python
ub
```

```python
lb
```

```python
## capping
```

```python
life[i]=np.where(life[i]>ub,np.quantile(life[i],.9),life[i])
```

```python
life[i]=np.where(life[i]<lb,np.quantile(life[i],.1),life[i])
```

```python
sns.boxplot(data=life,orient='h')
```

```python
outliers
```

```python
outliers.shape
```

```python
# outliers are replaced by the capping method where the values that are lower than the lower bound are replaced by the 10th percentile value and the values that lie above the upper bound are replaced by 90th percentile value
```

```python
x=life.drop(['Life_expectancy'],axis=1)
```

```python
y=life['Life_expectancy']
```

```python
life.columns
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(x,y,test_size=.2,random_state=100)
```

## transforming the features

```python
from sklearn.preprocessing import StandardScaler
```

```python
# create a StandardScaler object

scaler = StandardScaler()

# fit the scaler to the data and transform the data
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled=scaler.fit_transform(X_test)
```

# Random Forest Regressor

```python
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor(n_estimators=100,random_state=10,max_depth=5,min_samples_leaf=5) 
```

```python
model.fit(X_train_scaled,Y_train)
```

```python
Y_pred=model.predict(X_train_scaled)
```

```python
rf_resid=Y_train-Y_pred
```

```python
plt.plot (Y_pred, rf_resid, 'bo')
plt.ylabel ("Residuals")
plt.xlabel ("Predicted Strength")
plt.plot (Y_pred, 0*Y_pred)
```

residuals have uniform variance but there are few outliers to be treated

```python
# observed label vs predicted
plt.plot (Y_pred, Y_train, 'bo')
plt.ylabel ("Observed Strength")
plt.xlabel ("Predicted Strength")
plt.plot (Y_pred, Y_pred)
```

Model fitted well and is good at predicting the predictor variable

```python
Y_pred_tr1=model.predict(X_train_scaled)
```

```python
Y_pred_test1=model.predict(X_test_scaled)
```

```python
from sklearn.metrics import r2_score, mean_squared_error
```

```python
r2_score(Y_train,Y_pred_tr1)
```

```python
r2_score(Y_test,Y_pred_test1)
```

```python
rmse_tr1=np.sqrt(mean_squared_error(Y_train,Y_pred_tr1))
```

```python
rmse_test1=np.sqrt(mean_squared_error(Y_test,Y_pred_test1))
```

```python
rmse_tr1
```

```python
rmse_test1
```
